```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Financial Performance as dictionary
financial_performance = {
    "Month":["Jan","Feb","Mar","Apr","May"],
    "Revenue":[120, 135, 150, 160, 155],
    "Expenses":[80, 85, 95, 100, 98],
    "ProfitMargin":[33, 37, 35, 38, 37],
    "CustomerGrowth":[5, 6, 8, 9, 7],
    "MarketingIndex":[1.2, 1.3, 1.5, 1.6, 1.4]
}
df = pd.DataFrame(financial_performance)
print(df)

#Line plot for trends
plt.figure()
plt.plot(df["Month"], df["Revenue"], marker='x', label='Revenue')
plt.plot(df["Month"], df["Expenses"], marker='x', label='Expenses')
plt.title("FINANCIAL PERFORMANCE OVER TIME")
plt.xlabel("MONTH")
plt.ylabel("AMOUNT")
plt.legend()
plt.show()

#Scatter plot for relationship
plt.figure()
plt.plot(df["MarketingIndex"], df["CustomerGrowth"], marker='o')
plt.xlabel("Marketing Index")
plt.ylabel("Customer Growth")
plt.title("Trend of Marketing vs Growth")
plt.show()

# Correlation heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd

passenger_transport = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "PassengerCount": [112, 118, 132, 129, 121, 135],
    "AvgTemp": [23, 24, 26, 27, 28, 30],
    "Rainfall": [12, 8, 10, 9, 15, 11]
}

df = pd.DataFrame(passenger_transport )

# Rolling average
df["RollingAvg"] = df["PassengerCount"].rolling(window=2).mean()

# Percentage change
df["PctChange"] = df["PassengerCount"].pct_change() * 100

print(df)

# Variance & Std Dev
variance = df["PassengerCount"].var()
std_dev = df["PassengerCount"].std()

# Correlation
correlation = df.corr(numeric_only=True)

print("Variance:", variance)
print("Std Dev:", std_dev)
print("Correlation:\n", correlation)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # Import StandardScaler
from scipy.spatial.distance import cdist # Import cdist for distance calculation

features = {
    "Feature1": [5.1, 4.9, 6.2, 5.9, 5.6, 5.7],
    "Feature2": [3.5, 3.0, 3.4, 3.0, 2.9, 2.8],
    "Feature3": [1.4, 1.4, 5.4, 5.1, 3.6, 4.1],
    "Feature4": [0.2, 0.2, 2.3, 1.8, 1.3, 1.3]
}

df = pd.DataFrame(features)

# Calculate the distance matrix between all observations
dist_matrix = cdist(df, df, metric='euclidean')

# Compare distance between first and second observation vs first and third
d12 = dist_matrix[0, 1]
d13 = dist_matrix[0, 2]

print("Distance(Obs1, Obs2):", d12)
print("Distance(Obs1, Obs3):", d13)

if d12 < d13:
    print("Obs1 is closer to Obs2 than to Obs3")
else:
    print("Obs1 is closer to Obs3 than to Obs2")

# Distance between first two observations (this line was already correct for a single pair)
dist = np.linalg.norm(df.iloc[0] - df.iloc[1])
dist = round(dist, 2)
print("Distance:", dist)
```
